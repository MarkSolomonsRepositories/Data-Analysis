```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'C:/Life/Career and Degrees - REPO/B.M -Degree/Year2SemesterB/Data Anaysis/Exam prep/Tigbur files/BEFORE EXAM MAIN/AllData.xlsx'
Data = pd.read_excel(path,sheet_name='Employees')
Data.head()
```

```python
Data['FullName'] = Data['FirstName'] + Data['LastName']
Data = Data.loc[Data['MaritalStatus'] == 'S']
Data.head()
```

```python
result = Data.groupby(['Region Manager']).count()['FullName']
# result = data.groupby(['Region Manager']).count()['FullName']

result = pd.DataFrame(result).reset_index()

result
```

```python
sns.barplot(x='Region Manager',y='FullName',data=result)
plt.xticks(rotation=60)
```

```python
for i in Data['Region Manager']:
    output = Data.loc[Data['Region Manager']==i]
    output.to_csv('Olla' + str(i) + '.csv')
    
```
